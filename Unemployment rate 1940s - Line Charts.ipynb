## Improvement Suggestions:

While it is useful to look at plots in isolation, they appear to have varying y-axis values that make it hard to factor in their comparative variances. I feel it would be appropriate to evaluate them all after a normalisation, or perhaps all on the same plot.

```python
# Generating one plot to compare the yearly difference in unemployment rates: 
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(1,1,1)
ax2 = fig.add_subplot(1,1,1)
ax3 = fig.add_subplot(1,1,1)
ax4 = fig.add_subplot(1,1,1)
ax5 = fig.add_subplot(1,1,1)
ax1.plot(unrate[0:12]['MONTH'], unrate[0:12]['VALUE'], c='red', label='1948')

ax2.plot(unrate[12:24]['MONTH'], unrate[12:24]['VALUE'], c='blue', label='1949')

ax3.plot(unrate[24:36]['MONTH'], unrate[24:36]['VALUE'], c='green', label='1950')

ax4.plot(unrate[36:48]['MONTH'], unrate[36:48]['VALUE'], c='orange', label='1951')

ax5.plot(unrate[48:60]['MONTH'], unrate[48:60]['VALUE'], c='black', label='1952')

plt.legend(loc='upper left')
plt.show()
```

```python
# Refactored for x, y and title on single plt:

fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948 + i)
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=label)
    
plt.legend(loc='upper left')
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.xlabel("Month")
plt.ylabel("% Unemployment Rate")
plt.show()
```

## Findings:

Having explored the US unemployment rates (UR) between 1948 - 1952, the following insights have been discovered:

* There is a clear indication that 1949 saw the largest spike in UR of the 5 years.
* 1950 saw the best overall improvement over the 12 months having dropped from ~6.5% to ~4.4% after 12 months with a gradual decline in UR.
* 1948, 1951 and 1952, alll remained relatively consistent with their UR.
* 1950 and 1949 had the most variable fluctuations of their UR compared to all other years with 1949 seeing an increase of ~2.4% over the 12 months, hitting a peak of ~8% UR in the month of October.
* 1950 saw a return to lower UR with a gradual decline in UR of ~6.5% down to ~4.3% over 12 months, and a return to stability over the following 3 years there after with a gradual improvement as the years progressed.
* 1952 saw the lowest UR of each of the years with a min UR of ~2.8%.

It is expected that the rising UR in 1949 sparked action by the US government which lead to the gradual decline in UR in 1950 and the following years. 

This website has provided a good resource to explain the fluctuation of UR in the US since 1942: https://www.thebalance.com/unemployment-rate-by-year-3305506

Apparently the Korean war and min wage increase to $0.75 may be what has influenced the decrease in UR in 1950.
